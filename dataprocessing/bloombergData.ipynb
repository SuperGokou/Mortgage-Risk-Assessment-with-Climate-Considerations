```python
import blpapi
import pdblp
from xbbg import blp
import pandas as pd
import numpy as np
import datetime

```

```python
def get_data(tickers, fields, start_date, end_date):
    data = blp.bdh(tickers, fields, start_date, end_date)
    return pd.DataFrame(data)


start_date = '19940930'
end_date = datetime.datetime.today().strftime('%Y%m%d')

us_unemployment = get_data(['EHUPUS Index'], ['PX_LAST'], start_date, end_date)
display(us_unemployment)
```

```python
us_sale5aandbuy30yrs = get_data(['USSWIT10 BGN Curncy'] ['PX_LAST'], start_date, end_date)
display(us_sale5aandbuy30yrs)
```

```python
# 1st way to get data by blpapi
ticker = ['EHUPUS Index', 'USSWIT10 Curncy']
fields = ['PX_LAST']
start_date = '19940930'
# end_date = '20240131'

end_date = datetime.datetime.today().strftime('%Y%m%d')

data = blp.bdh(ticker, fields, start_date, end_date)

data = pd.DataFrame(data)
display(data)
```

```python
#2nd way to get data by pdblp
#change the port number to 8194 if you are using the desktop version of Bloomberg terminal and 8196 if you are using the Bloomberg Anywhere version
con = pdblp.BCon(debug=True, port=8194, timeout=5000)
con.start()

data = con.bdh(ticker, fields, start_date, end_date)
con.stop()

display(data)
```

```python
#save data to csv
data.to_csv('data.csv')
```
